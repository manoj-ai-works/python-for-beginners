# Operators
Operators perform operations on variables and values.

## 1. Arithmetic Operators

```python
a, b = 10, 3
print(a + b)   # 13  — addition
print(a - b)   # 7   — subtraction
print(a * b)   # 30  — multiplication
print(a / b)   # 3.333 — true division, always returns float
print(a // b)  # 3   — floor division, discards the decimal part
print(a % b)   # 1   — modulus: remainder after division
print(a ** b)  # 1000 — exponentiation: 10 to the power of 3
```

## 2. Comparison Operators

```python
x, y = 5, 10
print(x == y)  # False — checks if x equals y
print(x != y)  # True  — checks if x is not equal to y
print(x < y)   # True  — checks if x is less than y
print(x >= 5)  # True  — checks if x is greater than or equal to 5
```

## 3. Logical Operators

```python
print(True and False)  # False — both must be True for 'and' to return True
print(True or False)   # True  — at least one must be True for 'or' to return True
print(not True)        # False — 'not' flips the boolean value
age = 20
print(age >= 18 and age < 60)  # True — combine two conditions with 'and'
```

## 4. Assignment Operators

```python
n = 10
n += 5   # same as n = n + 5; shorthand to add and reassign
print(n)  # 15
n *= 2   # same as n = n * 2; shorthand to multiply and reassign
print(n)  # 30
n //= 4  # same as n = n // 4; floor divide and reassign
print(n)  # 7
```
