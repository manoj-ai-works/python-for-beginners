# Variables and Data Types
A variable is a name that holds a value. Python is dynamically typed.

## 1. Declaring Variables

```python
name = "Alice"      # store a text (string) value in a variable
age = 25            # store a whole number (integer)
height = 5.6        # store a decimal number (float)
is_student = True   # store a boolean (True or False)
print(name, age, height, is_student)  # print all four values separated by spaces
```

## 2. Data Types

```python
print(type(10))      # int   — whole numbers
print(type(3.14))    # float — decimal numbers
print(type('hello')) # str   — text
print(type(True))    # bool  — True or False
print(type(None))    # NoneType — represents the absence of a value
```

## 3. Type Conversion

```python
print(int('42'))     # convert string '42' to integer 42
print(float('3.14')) # convert string '3.14' to float 3.14
print(str(100))      # convert integer 100 to string '100'
print(bool(0))       # 0 is falsy, so this returns False
```

## 4. Multiple Assignment

```python
a, b, c = 1, 2, 3   # assign multiple variables in one line
print(a, b, c)
a, b = b, a          # swap two variables without a temporary variable
print(a, b)          # a is now 2, b is now 1
```
