# Strings
Strings are sequences of characters enclosed in quotes.

## 1. Creating Strings

```python
s1 = 'hello'          # single quotes work fine for strings
s2 = "world"          # double quotes are equivalent to single quotes
s3 = '''multi
line
string'''             # triple quotes allow strings that span multiple lines
print(s1, s2)
print(s3)
```

## 2. Indexing and Slicing

```python
s = "Python"
print(s[0])    # P      — index 0 is the first character
print(s[-1])   # n      — negative index counts from the end
print(s[1:4])  # yth    — slice from index 1 up to (not including) 4
print(s[::-1]) # nohtyP — step of -1 reverses the string
```

## 3. Common String Methods

```python
s = "  Hello, World!  "
print(s.strip())                   # remove leading and trailing whitespace
print(s.lower())                   # convert all characters to lowercase
print(s.upper())                   # convert all characters to uppercase
print(s.replace('World', 'Python'))# replace a substring with another
print(s.split(','))                # split into a list using ',' as separator
print(len(s))                      # number of characters including spaces
print('Hello' in s)                # check if a substring exists inside the string
```

## 4. f-Strings

```python
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")  # embed variables directly in the string
print(f"Next year I'll be {age + 1}.")                 # expressions inside {} are evaluated
```
