# User Input
`input()` reads a string from the user. Always returns a string — convert if needed.

```python
name = input("Enter your name: ")  # pause and wait for the user to type something
print(f"Hello, {name}!")           # use the entered value in an f-string
```

## Numeric Input

```python
age = int(input("Enter your age: "))  # input() always returns a string, so we convert to int
print(f"In 10 years, you'll be {age + 10}.")  # now we can do math with it
```

## Multiple Inputs

```python
a, b = input("Enter two numbers: ").split()  # split() breaks the input string on spaces
a, b = int(a), int(b)                        # convert both parts from string to integer
print(f"Sum = {a + b}")                      # now we can add them numerically
```
