# Modules and Imports
A module is a Python file with reusable code. Python has a rich standard library.

```python
import math              # import the entire math module from the standard library
print(math.pi)           # access the constant pi via the module name
print(math.sqrt(16))     # call sqrt() — must prefix with 'math.' when using import
print(math.ceil(4.2))    # ceil() rounds up to the nearest integer
```

```python
from random import randint, choice  # import specific functions — no prefix needed
print(randint(1, 100))              # generate a random integer between 1 and 100 (inclusive)
print(choice(['apple', 'banana', 'cherry']))  # pick a random item from the list
```

```python
import datetime                          # import the datetime module
now = datetime.datetime.now()            # get the current date and time
print(now)                               # full datetime object
print(now.strftime('%Y-%m-%d'))          # format as a readable date string
```
