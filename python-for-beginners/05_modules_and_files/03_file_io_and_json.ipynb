# File I/O and JSON
Python can read/write text files and serialize data to JSON — essential for saving and loading application state.

## 1. Writing and Reading Text Files

```python
# Write — 'w' mode creates the file or overwrites it
with open('notes.txt', 'w') as f:
    f.write('Line 1\n')                      # \n adds a newline so lines don't run together
    f.write('Line 2\n')
    f.writelines(['Line 3\n', 'Line 4\n'])   # writelines writes each item in the list

# Read all at once — loads the entire file into one string
with open('notes.txt', 'r') as f:
    content = f.read()
print(content)

# Read line by line — memory efficient for large files
with open('notes.txt', 'r') as f:
    for line in f:
        print(line.strip())  # strip() removes the trailing newline from each line
```

## 2. File Modes

| Mode | Meaning |
|---|---|
| `'r'` | Read (default) |
| `'w'` | Write (overwrites) |
| `'a'` | Append |
| `'x'` | Create (fails if exists) |
| `'rb'`/`'wb'` | Binary read/write |

## 3. JSON — Serialize Python Objects

```python
import json  # json module is part of the standard library

data = {
    'name': 'Alice',
    'age': 25,
    'scores': [90, 85, 92],
    'active': True
}

# Write JSON to file — json.dump() converts Python object to JSON and writes it
with open('data.json', 'w') as f:
    json.dump(data, f, indent=2)  # indent=2 makes the file human-readable

# Read JSON from file — json.load() reads JSON and converts it back to a Python object
with open('data.json', 'r') as f:
    loaded = json.load(f)

print(loaded)
print(type(loaded))  # dict — JSON objects become Python dicts
```

## 4. JSON with Lists of Objects

```python
users = [
    {'id': 1, 'name': 'Alice', 'email': 'alice@example.com'},
    {'id': 2, 'name': 'Bob',   'email': 'bob@example.com'},
]

with open('users.json', 'w') as f:
    json.dump(users, f, indent=2)  # a list of dicts serializes to a JSON array

with open('users.json', 'r') as f:
    loaded_users = json.load(f)    # JSON array becomes a Python list of dicts

for user in loaded_users:
    print(f"{user['id']}: {user['name']}")
```

## 5. json.dumps / json.loads — String conversion

```python
# json.dumps — convert Python object to a JSON string (not a file)
json_str = json.dumps({'key': 'value', 'num': 42}, indent=2)
print(json_str)
print(type(json_str))  # str — useful for sending data over a network

# json.loads — parse a JSON string back into a Python object
parsed = json.loads(json_str)
print(parsed['key'])   # 'value' — now it's a regular Python dict
```

## 6. Handling Missing Files Safely

```python
import os

def load_data(path: str) -> list:
    if not os.path.exists(path):  # check if the file exists before trying to open it
        return []                 # return empty list instead of crashing
    try:
        with open(path, 'r') as f:
            return json.load(f)
    except json.JSONDecodeError:  # handle corrupted or invalid JSON gracefully
        print('Warning: corrupted file')
        return []

result = load_data('nonexistent.json')
print(result)  # []
```
