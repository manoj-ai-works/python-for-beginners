# File Handling
Python can read and write files using the built-in `open()` function.

## 1. Writing a File

```python
with open('sample.txt', 'w') as f:  # 'w' mode creates the file or overwrites it if it exists
    f.write('Hello, World!\n')      # write a line; \n adds a newline at the end
    f.write('Python file handling is easy.')  # write a second line
```

## 2. Reading a File

```python
with open('sample.txt', 'r') as f:  # 'r' mode opens for reading (default mode)
    content = f.read()              # read the entire file content as one string
print(content)
```

## 3. Reading Line by Line

```python
with open('sample.txt', 'r') as f:
    for line in f:              # iterate over the file object — memory efficient for large files
        print(line.strip())     # strip() removes the trailing newline character from each line
```

## 4. Appending

```python
with open('sample.txt', 'a') as f:  # 'a' mode adds to the end without overwriting existing content
    f.write('\nAppended line.')
```

## File Modes
- `'r'` — read (default)
- `'w'` — write (overwrites)
- `'a'` — append
- `'rb'` / `'wb'` — binary mode
