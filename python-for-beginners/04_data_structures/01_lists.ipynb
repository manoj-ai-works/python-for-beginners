# Lists
Lists are ordered, mutable collections that allow duplicates.

```python
fruits = ['apple', 'banana', 'cherry']
print(fruits[0])   # apple  — index 0 is the first element
print(fruits[-1])  # cherry — negative index counts from the end
print(fruits[1:3]) # ['banana', 'cherry'] — slice from index 1 up to (not including) 3
```

## Common Methods

```python
nums = [3, 1, 4, 1, 5, 9]
nums.append(2)       # add a new element to the end of the list
nums.insert(0, 0)    # insert value 0 at index 0 (shifts everything right)
nums.remove(1)       # remove the first occurrence of value 1
nums.sort()          # sort the list in place (modifies the original list)
print(nums)
print(len(nums))     # number of elements in the list
print(nums.count(1)) # count how many times 1 appears in the list
```

## List Operations

```python
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)    # concatenate: join two lists into one
print(a * 2)    # repeat: create a new list with 'a' repeated twice
print(3 in a)   # membership test: check if 3 exists in the list
```
