# Dictionaries
Dictionaries store key-value pairs. Keys must be unique and immutable.

```python
person = {'name': 'Alice', 'age': 25, 'city': 'NYC'}
print(person['name'])              # Alice — access value by key (raises KeyError if missing)
print(person.get('age'))           # 25    — .get() is safer: returns None if key not found
print(person.get('email', 'N/A'))  # N/A   — .get() with a default value avoids KeyError
```

## CRUD Operations

```python
d = {}
d['key'] = 'value'   # create a new key or update an existing one
print(d['key'])       # read the value associated with 'key'
del d['key']          # delete the key-value pair from the dictionary
print(d)              # {} — empty dict after deletion
```

## Iterating

```python
scores = {'Alice': 90, 'Bob': 85, 'Charlie': 92}

for key in scores:              # iterating a dict gives you the keys by default
    print(key)

for key, value in scores.items():  # .items() gives (key, value) pairs for both at once
    print(f"{key}: {value}")
```
