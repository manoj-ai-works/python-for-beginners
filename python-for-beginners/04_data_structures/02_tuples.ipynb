# Tuples
Tuples are ordered, immutable collections. Use them for data that shouldn't change.

```python
point = (3, 4)       # create a tuple with two values
print(point[0])      # 3 — access elements by index, just like a list

# Unpacking — assign each element to a separate variable in one step
x, y = point
print(x, y)          # 3 4 — x gets 3, y gets 4
```

## Tuple vs List

```python
t = (1, 2, 3)
# t[0] = 10  # TypeError: tuples are immutable — you cannot change their elements

# But you can create a new tuple by concatenating
t2 = t + (4,)  # (4,) is a one-element tuple — the comma is required
print(t2)      # (1, 2, 3, 4) — original t is unchanged
```

## Named Tuples

```python
from collections import namedtuple  # import the namedtuple factory from the standard library

Point = namedtuple('Point', ['x', 'y'])  # create a tuple class with named fields
p = Point(3, 4)      # create an instance just like a regular tuple
print(p.x, p.y)      # access fields by name instead of index — much more readable
```
