# Sets
Sets are unordered collections of unique elements. Great for removing duplicates and set operations.

```python
s = {1, 2, 3, 2, 1}  # duplicate values are automatically removed
print(s)              # {1, 2, 3} — order is not guaranteed in sets

s.add(4)              # add a new element (no effect if it already exists)
s.remove(2)           # remove element 2 (raises KeyError if not found)
print(s)
```

## Set Operations

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)  # union: all elements from both sets (no duplicates)
print(a & b)  # intersection: only elements that appear in BOTH sets
print(a - b)  # difference: elements in 'a' that are NOT in 'b'
print(a ^ b)  # symmetric difference: elements in one set but not both
```

## Remove Duplicates from List

```python
nums = [1, 2, 2, 3, 3, 3, 4]
unique = list(set(nums))  # convert to set removes duplicates, then back to list
print(unique)             # note: order may differ from the original list
```
