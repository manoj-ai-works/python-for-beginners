# Dunder (Magic) Methods
Dunder methods like `__init__`, `__str__`, `__len__` let you define how objects behave with built-in operations.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x  # store the x component
        self.y = y  # store the y component

    def __str__(self):
        return f"Vector({self.x}, {self.y})"  # called by print() — human-readable output

    def __repr__(self):
        return f"Vector(x={self.x}, y={self.y})"  # called in the REPL — useful for debugging

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)  # enables v1 + v2 syntax

    def __len__(self):
        return int((self.x**2 + self.y**2) ** 0.5)  # enables len(v) — returns magnitude as int

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y  # enables v1 == v2 comparison


v1 = Vector(2, 3)
v2 = Vector(1, 4)
print(v1)          # Vector(2, 3)  — calls __str__
print(v1 + v2)     # Vector(3, 7)  — calls __add__
print(len(v1))     # 3             — calls __len__
print(v1 == v2)    # False         — calls __eq__
```

## Common Dunder Methods
| Method | Triggered by |
|---|---|
| `__init__` | `ClassName()` |
| `__str__` | `print(obj)` |
| `__len__` | `len(obj)` |
| `__add__` | `obj1 + obj2` |
| `__eq__` | `obj1 == obj2` |
| `__lt__` | `obj1 < obj2` |
| `__getitem__` | `obj[key]` |
