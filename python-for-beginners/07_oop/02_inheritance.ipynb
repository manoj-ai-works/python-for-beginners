# Inheritance
Inheritance lets a class reuse and extend another class.

```python
class Animal:
    def __init__(self, name):
        self.name = name  # base class stores the name shared by all animals

    def speak(self):
        return f"{self.name} makes a sound."  # default behavior — subclasses can override this


class Dog(Animal):          # Dog inherits everything from Animal
    def speak(self):        # override the parent's speak() with Dog-specific behavior
        return f"{self.name} says Woof!"


class Cat(Animal):
    def speak(self):        # each subclass provides its own version of speak()
        return f"{self.name} says Meow!"


for animal in [Dog('Rex'), Cat('Whiskers')]:
    print(animal.speak())  # Python calls the correct speak() based on the actual type
```

## super()

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed


class Car(Vehicle):
    def __init__(self, brand, speed, doors):
        super().__init__(brand, speed)  # call the parent __init__ to set brand and speed
        self.doors = doors              # then add the Car-specific attribute

    def info(self):
        return f"{self.brand}, {self.speed}km/h, {self.doors} doors"


c = Car('Toyota', 180, 4)
print(c.info())
```
