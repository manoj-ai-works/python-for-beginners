# Encapsulation and Polymorphism

## Encapsulation
Restricting direct access to data using private/protected attributes.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # double underscore makes this a private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount  # only the class itself can modify __balance

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")  # guard against invalid withdrawals

    def get_balance(self):
        return self.__balance  # provide controlled read access via a public method


acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.get_balance())  # 1300 — access balance through the getter, not directly
# print(acc.__balance)   # AttributeError — private attribute is not accessible from outside
```

## Polymorphism
Same interface, different behavior depending on the object.

```python
class Circle:
    def __init__(self, r): self.r = r
    def area(self): return 3.14 * self.r ** 2  # area formula for a circle

class Rectangle:
    def __init__(self, w, h): self.w = w; self.h = h
    def area(self): return self.w * self.h     # area formula for a rectangle

shapes = [Circle(5), Rectangle(4, 6)]
for shape in shapes:
    print(shape.area())  # same .area() call works on both — Python picks the right method
```
