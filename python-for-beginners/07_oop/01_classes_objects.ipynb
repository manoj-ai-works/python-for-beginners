# Classes and Objects
A class is a blueprint. An object is an instance of that blueprint.

```python
class Dog:
    # Class attribute — shared by ALL instances of Dog
    species = 'Canis familiaris'

    def __init__(self, name, age):  # __init__ is called automatically when creating an object
        # Instance attributes — unique to each Dog object
        self.name = name  # 'self' refers to the specific instance being created
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"  # access instance attributes via self

    def __str__(self):
        return f"Dog(name={self.name}, age={self.age})"  # controls what print(dog) shows


dog1 = Dog('Rex', 3)    # create an instance — __init__ runs with name='Rex', age=3
dog2 = Dog('Buddy', 5)  # create a second independent instance

print(dog1.bark())   # Rex says Woof!
print(dog2)          # calls __str__: Dog(name=Buddy, age=5)
print(Dog.species)   # access class attribute via the class name
```

## Class vs Instance Attributes

```python
class Counter:
    count = 0  # class attribute — shared across all instances

    def __init__(self):
        Counter.count += 1  # increment the shared counter each time a new object is created

c1 = Counter()  # count becomes 1
c2 = Counter()  # count becomes 2
print(Counter.count)  # 2 — both instances share the same class attribute
```
