# Dataclasses
`@dataclass` auto-generates `__init__`, `__repr__`, `__eq__` for you. Less boilerplate, cleaner code.

## 1. Basic Dataclass

```python
from dataclasses import dataclass  # import the decorator from the standard library

@dataclass                          # this decorator auto-generates __init__, __repr__, __eq__
class Point:
    x: float  # type annotations are required — they tell @dataclass what fields to create
    y: float

p1 = Point(1.0, 2.0)  # __init__ is auto-generated, no need to write it manually
p2 = Point(1.0, 2.0)

print(p1)          # Point(x=1.0, y=2.0)  — __repr__ is auto-generated
print(p1 == p2)    # True                 — __eq__ compares field values automatically
p1.x = 10          # fields are mutable by default — you can change them
```

## 2. Default Values and field()

```python
from dataclasses import dataclass, field

@dataclass
class Student:
    name: str
    grade: str = 'A'                          # simple default value
    scores: list = field(default_factory=list) # mutable defaults MUST use field() to avoid sharing

    def add_score(self, score: int):
        self.scores.append(score)  # safely append to this instance's own list

s = Student('Alice')  # grade defaults to 'A', scores defaults to a new empty list
s.add_score(95)
s.add_score(88)
print(s)
```

## 3. frozen=True — Immutable Dataclass

```python
@dataclass(frozen=True)  # frozen=True makes all fields read-only after creation
class Color:
    r: int
    g: int
    b: int

    def __str__(self):
        return f'rgb({self.r}, {self.g}, {self.b})'

red = Color(255, 0, 0)
print(red)
print(hash(red))  # frozen dataclasses are hashable — can be used as dict keys or in sets

try:
    red.r = 100   # raises FrozenInstanceError — frozen objects cannot be modified
except Exception as e:
    print(e)
```

## 4. __post_init__ — Validation

```python
@dataclass
class Temperature:
    celsius: float

    def __post_init__(self):  # called automatically after __init__ — perfect for validation
        if self.celsius < -273.15:
            raise ValueError(f'Temperature {self.celsius} below absolute zero!')

    @property
    def fahrenheit(self) -> float:
        return self.celsius * 9/5 + 32  # computed property — not stored, calculated on access

t = Temperature(100)
print(t.fahrenheit)  # 212.0

try:
    Temperature(-300)  # triggers __post_init__ validation
except ValueError as e:
    print(e)
```

## 5. Inheritance with Dataclasses

```python
@dataclass
class Animal:
    name: str
    sound: str

@dataclass
class Dog(Animal):          # inherit fields from Animal and add new ones
    breed: str
    sound: str = 'Woof'     # override the default value for 'sound'

d = Dog(name='Rex', breed='Labrador')  # sound uses the default 'Woof'
print(d)
```
