# break, continue, pass
These keywords control loop execution flow.

## 1. break — exit the loop

```python
for i in range(10):
    if i == 5:
        break       # immediately exit the loop when i reaches 5
    print(i)        # prints 0 to 4 — never reaches 5 or beyond
```

## 2. continue — skip current iteration

```python
for i in range(10):
    if i % 2 == 0:
        continue    # skip the rest of this iteration and go to the next one
    print(i)        # only odd numbers are printed because evens are skipped
```

## 3. pass — do nothing (placeholder)

```python
for i in range(5):
    if i == 3:
        pass        # pass does nothing — useful as a placeholder for future code
    print(i)        # all numbers still print because pass doesn't skip or stop
```
