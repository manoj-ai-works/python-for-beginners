# If / Else
Control flow lets your program make decisions.

## 1. Basic if-else

```python
age = 18
if age >= 18:       # check if the condition is True
    print("Adult")  # this block runs when the condition is True
else:
    print("Minor")  # this block runs when the condition is False
```

## 2. elif

```python
score = 75
if score >= 90:      # check the first condition
    print("A")
elif score >= 80:    # only checked if the previous condition was False
    print("B")
elif score >= 70:    # checked only if all previous conditions were False
    print("C")
else:
    print("F")       # fallback if none of the above conditions matched
```

## 3. Ternary (one-liner)

```python
x = 10
# ternary: value_if_true if condition else value_if_false
result = "even" if x % 2 == 0 else "odd"  # compact way to write a simple if-else
print(result)
```

## 4. Nested if

```python
num = 15
if num > 0:              # outer condition: is the number positive?
    if num % 2 == 0:     # inner condition: is it also even?
        print("Positive even")
    else:
        print("Positive odd")  # positive but not even
else:
    print("Non-positive")      # num is 0 or negative
```
