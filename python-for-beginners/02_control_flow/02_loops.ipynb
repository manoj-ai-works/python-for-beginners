# Loops
Loops repeat a block of code multiple times.

## 1. for Loop

```python
for i in range(5):  # range(5) produces 0, 1, 2, 3, 4 — loop runs 5 times
    print(i)        # prints 0 1 2 3 4
```

```python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:  # iterate directly over each item in the list
    print(fruit)      # fruit takes the value of each element in turn
```

## 2. while Loop

```python
count = 0
while count < 5:   # keep looping as long as this condition is True
    print(count)
    count += 1     # increment count each iteration to avoid an infinite loop
```

## 3. range()

```python
# range(start, stop, step) — start is inclusive, stop is exclusive
for i in range(1, 10, 2):  # start at 1, stop before 10, step by 2
    print(i)               # prints 1 3 5 7 9
```

## 4. enumerate() and zip()

```python
names = ['Alice', 'Bob', 'Charlie']
for i, name in enumerate(names):  # enumerate gives both index and value
    print(i, name)

scores = [90, 85, 92]
for name, score in zip(names, scores):  # zip pairs elements from two lists together
    print(f"{name}: {score}")
```
