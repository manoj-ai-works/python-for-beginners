# Unit Testing
Unit tests verify individual pieces of code work correctly. Python's built-in `unittest` module handles this.

## 1. Basic Test Structure

```python
import unittest

def add(a, b): return a + b
def divide(a, b):
    if b == 0: raise ValueError('Cannot divide by zero')  # guard against division by zero
    return a / b

class TestMath(unittest.TestCase):  # test class must inherit from unittest.TestCase

    def test_add(self):             # each test method must start with 'test_'
        self.assertEqual(add(2, 3), 5)   # assert the result equals the expected value
        self.assertEqual(add(-1, 1), 0)  # test with negative numbers too

    def test_divide(self):
        self.assertEqual(divide(10, 2), 5.0)  # verify normal division works

    def test_divide_by_zero(self):
        with self.assertRaises(ValueError):  # assert that a ValueError is raised
            divide(10, 0)

unittest.main(argv=[''], exit=False, verbosity=2)  # run tests in Jupyter without exiting
```

## 2. Common Assertions

| Method | Checks |
|---|---|
| `assertEqual(a, b)` | a == b |
| `assertNotEqual(a, b)` | a != b |
| `assertTrue(x)` | bool(x) is True |
| `assertFalse(x)` | bool(x) is False |
| `assertIsNone(x)` | x is None |
| `assertIn(a, b)` | a in b |
| `assertRaises(Exc)` | exception is raised |

## 3. setUp and tearDown

```python
class TestWithSetup(unittest.TestCase):

    def setUp(self):
        self.data = [1, 2, 3, 4, 5]  # setUp runs before EACH test — provides a fresh state

    def tearDown(self):
        self.data = None              # tearDown runs after EACH test — clean up resources

    def test_sum(self):
        self.assertEqual(sum(self.data), 15)  # self.data is always fresh from setUp

    def test_length(self):
        self.assertEqual(len(self.data), 5)

unittest.main(argv=[''], exit=False, verbosity=2)
```
