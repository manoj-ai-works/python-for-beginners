# Exception Handling
Exceptions are errors that occur at runtime. Use `try/except` to handle them gracefully.

## 1. Basic try/except

```python
try:
    x = int(input("Enter a number: "))  # this could raise ValueError if input is not a number
    print(10 / x)                       # this could raise ZeroDivisionError if x is 0
except ZeroDivisionError:
    print("Cannot divide by zero!")     # handle the specific case of dividing by zero
except ValueError:
    print("That's not a valid number!") # handle the case where conversion to int fails
```

## 2. else and finally

```python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error!")
else:
    print(f"Result: {result}")  # 'else' runs only if NO exception was raised in 'try'
finally:
    print("This always runs.")  # 'finally' runs no matter what — great for cleanup code
```

## 3. Raising Exceptions

```python
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")  # manually trigger an exception with a message
    return age

try:
    set_age(-5)
except ValueError as e:  # 'as e' captures the exception object so we can print its message
    print(e)
```

## 4. Custom Exceptions

```python
class InsufficientFundsError(Exception):  # inherit from Exception to create a custom error type
    pass                                   # 'pass' means no extra behavior — just a new name

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError("Not enough balance")  # raise our custom exception
    return balance - amount

try:
    withdraw(100, 200)
except InsufficientFundsError as e:
    print(e)
```
