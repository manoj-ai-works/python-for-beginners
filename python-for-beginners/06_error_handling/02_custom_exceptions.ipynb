# Custom Exceptions
Python lets you define your own exception classes by inheriting from `Exception`. This makes errors more descriptive and easier to catch specifically.

## 1. Why Custom Exceptions?

```python
# Generic — hard to know what went wrong
try:
    raise ValueError('something failed')  # ValueError is too vague — which validation failed?
except ValueError:
    print('caught a ValueError — but which one?')
```

## 2. Defining Custom Exceptions

```python
class AppError(Exception):       # base exception for this app — all custom errors inherit from this
    """Base exception for this application."""
    pass

class ValidationError(AppError): # specific error for input validation failures
    """Raised when input validation fails."""
    pass

class NotFoundError(AppError):   # specific error when a resource doesn't exist
    """Raised when a resource is not found."""
    pass

class DuplicateError(AppError):  # specific error for duplicate entries
    """Raised when a duplicate entry is detected."""
    pass
```

## 3. Raising and Catching

```python
def find_user(user_id: int, users: dict):
    if user_id not in users:
        raise NotFoundError(f'User {user_id} not found')  # raise with a descriptive message
    return users[user_id]

users = {1: 'Alice', 2: 'Bob'}

try:
    print(find_user(1, users))   # Alice — found successfully
    print(find_user(99, users))  # raises NotFoundError
except NotFoundError as e:
    print(f'Error: {e}')         # catch only this specific error type
```

## 4. Exception Hierarchy — catch base to handle all

```python
def process(value):
    if not isinstance(value, str):
        raise ValidationError('Must be a string')       # wrong type
    if value in ['admin', 'root']:
        raise DuplicateError(f"'{value}' is reserved")  # reserved name
    return value.upper()

for val in ['hello', 123, 'admin']:
    try:
        print(process(val))
    except AppError as e:  # catches ALL our custom exceptions because they all inherit AppError
        print(f'AppError: {e}')
```

## 5. Adding Extra Info to Exceptions

```python
class InsufficientFundsError(Exception):
    def __init__(self, balance: float, amount: float):
        self.balance = balance  # store extra data on the exception for the caller to use
        self.amount  = amount
        super().__init__(f'Cannot withdraw {amount}. Balance is only {balance}.')  # set the message

try:
    raise InsufficientFundsError(balance=50.0, amount=100.0)
except InsufficientFundsError as e:
    print(e)                              # prints the message from __init__
    print(f'Short by: {e.amount - e.balance}')  # access the stored attributes directly
```
