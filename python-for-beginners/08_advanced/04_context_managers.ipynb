# Context Managers
Context managers handle setup and teardown automatically using `with`. The most common example is file handling.

```python
# Without context manager (manual cleanup)
f = open('test.txt', 'w')
f.write('hello')
f.close()  # easy to forget — if an error occurs before this, the file stays open!

# With context manager (auto cleanup)
with open('test.txt', 'w') as f:
    f.write('hello')  # file closes automatically when the 'with' block exits, even on error
```

## Custom Context Manager with class

```python
class Timer:
    import time

    def __enter__(self):       # called when entering the 'with' block
        import time
        self.start = time.time()  # record start time
        return self            # the value assigned to 'as' variable

    def __exit__(self, *args): # called when leaving the 'with' block (even on exception)
        import time
        self.elapsed = time.time() - self.start
        print(f"Elapsed: {self.elapsed:.4f}s")

with Timer():  # __enter__ runs, then the block, then __exit__
    total = sum(range(1_000_000))
```

## Custom Context Manager with contextlib

```python
from contextlib import contextmanager  # decorator that turns a generator into a context manager

@contextmanager
def managed_resource(name):
    print(f"Acquiring {name}")  # setup code — runs before the 'with' block
    yield name                  # pause here and give control to the 'with' block
    print(f"Releasing {name}")  # teardown code — runs after the 'with' block exits

with managed_resource('database connection') as res:
    print(f"Using {res}")  # 'res' is the value passed to yield
```
