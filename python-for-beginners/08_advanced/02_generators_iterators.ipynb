# Generators and Iterators
Generators produce values lazily — one at a time — saving memory.

## Generator Function

```python
def count_up(n):
    i = 0
    while i < n:
        yield i  # 'yield' pauses the function and returns i; resumes from here on next call
        i += 1   # this runs after the caller requests the next value

for val in count_up(5):  # the for loop calls next() automatically each iteration
    print(val)
```

## Generator Expression

```python
# Like list comprehension but with () — values are generated on demand, not stored in memory
gen = (x**2 for x in range(10))
print(next(gen))  # 0 — manually request the first value
print(next(gen))  # 1 — request the second value
print(list(gen))  # consume all remaining values into a list
```

## Custom Iterator

```python
class Countdown:
    def __init__(self, n):
        self.n = n

    def __iter__(self):
        return self  # the object itself is the iterator

    def __next__(self):
        if self.n <= 0:
            raise StopIteration  # signal that there are no more values to produce
        self.n -= 1
        return self.n + 1  # return current value before decrement

for i in Countdown(5):  # Python calls __next__ each iteration until StopIteration
    print(i)
```
