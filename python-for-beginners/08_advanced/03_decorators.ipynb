# Decorators
A decorator wraps a function to add behavior without modifying it.

```python
def my_decorator(func):          # a decorator is a function that takes another function
    def wrapper(*args, **kwargs): # wrapper accepts any arguments so it works with any function
        print("Before function call")
        result = func(*args, **kwargs)  # call the original function and capture its return value
        print("After function call")
        return result            # return the original result so the caller gets it
    return wrapper               # return the wrapper, not the result — this replaces the function

@my_decorator                    # @syntax is shorthand for: say_hello = my_decorator(say_hello)
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
```

## Practical Examples

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()              # record the time before the function runs
        result = func(*args, **kwargs)
        end = time.time()                # record the time after it finishes
        print(f"{func.__name__} took {end - start:.4f}s")  # print elapsed time
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.1)  # simulate a slow operation
    return "done"

print(slow_function())
```

```python
def require_positive(func):
    def wrapper(n):
        if n < 0:
            raise ValueError("Must be positive")  # guard the function before it even runs
        return func(n)  # only call the original function if validation passes
    return wrapper

@require_positive
def square_root(n):
    return n ** 0.5

print(square_root(9))  # 3.0 — validation passes, function runs normally
```
