# Comprehensions — List, Dict, Set
Comprehensions are concise, readable ways to build collections from iterables.

## 1. List Comprehension

```python
# Basic — [expression for item in iterable]
squares = [x**2 for x in range(10)]  # square each number from 0 to 9
print(squares)

# With condition — only include items that pass the filter
evens = [x for x in range(20) if x % 2 == 0]
print(evens)

# Transform + filter — both transform and filter in one expression
words = ['hello', 'world', 'python', 'is', 'fun']
long_upper = [w.upper() for w in words if len(w) > 3]  # uppercase only words longer than 3
print(long_upper)
```

## 2. Dict Comprehension

```python
# Basic — {key: value for item in iterable}
word_lengths = {w: len(w) for w in ['apple', 'banana', 'cherry']}  # map word to its length
print(word_lengths)

# Invert a dict — swap keys and values
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}  # values become keys, keys become values
print(inverted)

# Filter while building — only include entries that meet a condition
scores = {'Alice': 90, 'Bob': 55, 'Charlie': 78, 'Diana': 45}
passing = {name: score for name, score in scores.items() if score >= 60}  # only passing scores
print(passing)
```

## 3. Set Comprehension

```python
# Unique lengths — set automatically removes duplicates
words = ['hi', 'hello', 'hey', 'world', 'wow']
unique_lengths = {len(w) for w in words}  # 'hi' and 'hey' both have length 2, stored once
print(unique_lengths)  # {2, 3, 5}

# Remove duplicates while transforming
nums = [1, -1, 2, -2, 3, 3]
abs_unique = {abs(n) for n in nums}  # abs() makes negatives positive; set removes duplicates
print(abs_unique)  # {1, 2, 3}
```

## 4. Nested Comprehension

```python
# Flatten a 2D list — outer loop first, then inner loop
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [n for row in matrix for n in row]  # read as: for each row, for each n in that row
print(flat)

# Multiplication table as dict — keys are (i, j) tuples
table = {(i, j): i * j for i in range(1, 4) for j in range(1, 4)}
for k, v in table.items():
    print(f'{k[0]} x {k[1]} = {v}')
```

## 5. Generator Expression (lazy version)

```python
# Uses () instead of [] — values are computed one at a time, not all stored in memory at once
total = sum(x**2 for x in range(1000000))  # efficient: no list of 1M items is ever created
print(total)
```
