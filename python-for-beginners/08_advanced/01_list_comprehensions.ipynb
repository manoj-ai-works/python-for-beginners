# List Comprehensions
A concise way to create lists. Faster and more readable than loops.

```python
# Basic — [expression for item in iterable]
squares = [x**2 for x in range(10)]  # compute x squared for each x from 0 to 9
print(squares)

# With condition — [expression for item in iterable if condition]
evens = [x for x in range(20) if x % 2 == 0]  # only include x when it's even
print(evens)

# Nested — outer loop first, then inner loop
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]  # 3x3 multiplication table
for row in matrix:
    print(row)
```

## Dict and Set Comprehensions

```python
# Dict comprehension — {key: value for item in iterable}
words = ['hello', 'world', 'python']
word_len = {w: len(w) for w in words}  # map each word to its length
print(word_len)

# Set comprehension — {expression for item in iterable} — duplicates are removed automatically
unique_squares = {x**2 for x in [-2, -1, 0, 1, 2]}  # -2 and 2 both give 4, so only one 4
print(unique_squares)
```
