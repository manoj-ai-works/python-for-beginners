# Lambda and Functional Tools
Lambda is an anonymous one-line function. Combined with `map`, `filter`, `sorted`, and `functools` it enables functional-style programming.

## 1. Lambda Basics

```python
square  = lambda x: x ** 2           # anonymous function: takes x, returns x squared
add     = lambda a, b: a + b         # lambdas can take multiple parameters
greater = lambda a, b: a if a > b else b  # ternary expression inside a lambda

print(square(5))       # 25
print(add(3, 4))       # 7
print(greater(10, 7))  # 10
```

## 2. map() — apply to every element

```python
nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))  # apply the lambda to each element; wrap in list()
print(squares)  # [1, 4, 9, 16, 25]

names = ['alice', 'bob', 'charlie']
capitalized = list(map(str.capitalize, names))  # pass a method directly instead of a lambda
print(capitalized)
```

## 3. filter() — keep matching elements

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8]
evens = list(filter(lambda x: x % 2 == 0, nums))  # keep only elements where lambda returns True
print(evens)  # [2, 4, 6, 8]

words = ['apple', 'hi', 'banana', 'ok', 'cherry']
long_words = list(filter(lambda w: len(w) > 3, words))  # keep words longer than 3 characters
print(long_words)
```

## 4. sorted() with key

```python
# Sort by length — key function returns the value to sort by
words = ['banana', 'apple', 'cherry', 'fig']
print(sorted(words, key=lambda w: len(w)))  # shortest first

# Sort list of dicts by a specific field
people = [{'name': 'Charlie', 'age': 30}, {'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 35}]
print(sorted(people, key=lambda p: p['age']))  # youngest first

# Sort tuples by second element
pairs = [(1, 'b'), (3, 'a'), (2, 'c')]
print(sorted(pairs, key=lambda x: x[1]))  # alphabetical by the second element
```

## 5. functools.reduce()

```python
from functools import reduce  # reduce is not a built-in, must import from functools

nums = [1, 2, 3, 4, 5]
# reduce applies the function cumulatively: ((((1*2)*3)*4)*5) = 120
product = reduce(lambda acc, x: acc * x, nums)
print(product)  # 120

# Find max using reduce — compare pairs and keep the larger one
max_val = reduce(lambda a, b: a if a > b else b, nums)
print(max_val)  # 5
```
