# Regular Expressions (Regex)
Regex lets you search, match, and validate patterns in strings using the `re` module.

## 1. Basic Matching

```python
import re  # re is the standard library module for regular expressions

text = 'Hello, my email is alice@example.com'
# \w+ matches one or more word characters; \. matches a literal dot
match = re.search(r'\w+@\w+\.\w+', text)  # r'' prefix = raw string, so \ is not escaped
if match:
    print(match.group())  # .group() returns the matched text: alice@example.com
```

## 2. Common Patterns

| Pattern | Meaning |
|---|---|
| `\d` | digit 0-9 |
| `\w` | word char (a-z, A-Z, 0-9, _) |
| `\s` | whitespace |
| `.` | any char except newline |
| `^` | start of string |
| `$` | end of string |
| `+` | one or more |
| `*` | zero or more |
| `?` | zero or one |
| `{n}` | exactly n times |

## 3. findall, sub, split

```python
text = 'Call 123-456-7890 or 987-654-3210'

# findall — returns a list of ALL matches in the string
phones = re.findall(r'\d{3}-\d{3}-\d{4}', text)  # \d{3} means exactly 3 digits
print(phones)

# sub — replace all matches with a new string
cleaned = re.sub(r'\d{3}-\d{3}-\d{4}', '[REDACTED]', text)  # hide phone numbers
print(cleaned)

# split — split the string at each match
parts = re.split(r'\s+', 'one  two   three')  # \s+ matches one or more whitespace chars
print(parts)
```

## 4. Validation Examples

```python
def is_valid_email(email):
    # ^ anchors to start, $ anchors to end — ensures the whole string matches
    return bool(re.match(r'^[\w.-]+@[\w.-]+\.\w{2,}$', email))

def is_valid_phone(phone):
    # \+? means optional +, [\d\s\-] matches digits, spaces, or dashes
    return bool(re.match(r'^\+?[\d\s\-]{7,15}$', phone))

print(is_valid_email('alice@example.com'))  # True
print(is_valid_email('not-an-email'))       # False
print(is_valid_phone('+1 800-555-1234'))    # True
```
