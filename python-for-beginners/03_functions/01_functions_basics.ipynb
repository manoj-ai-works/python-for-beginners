# Functions
Functions are reusable blocks of code defined with `def`.

## 1. Defining and Calling

```python
def greet(name):              # define a function called 'greet' that takes one parameter
    print(f"Hello, {name}!") # the body runs every time the function is called

greet("Alice")  # call the function and pass 'Alice' as the argument
greet("Bob")    # call it again with a different argument — reuse without rewriting
```

## 2. Return Values

```python
def add(a, b):
    return a + b  # 'return' sends a value back to the caller

result = add(3, 4)  # the returned value is stored in 'result'
print(result)       # 7
```

## 3. Default Parameters

```python
def greet(name, greeting="Hello"):  # 'greeting' has a default value used when not provided
    print(f"{greeting}, {name}!")

greet("Alice")           # uses default greeting: Hello, Alice!
greet("Bob", "Hi")       # overrides the default: Hi, Bob!
```

## 4. Multiple Return Values

```python
def min_max(numbers):
    return min(numbers), max(numbers)  # Python returns a tuple when you list multiple values

lo, hi = min_max([3, 1, 9, 2, 7])  # unpack the returned tuple into two variables
print(lo, hi)                       # 1 9
```
