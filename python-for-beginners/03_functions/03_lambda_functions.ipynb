# Lambda Functions
A lambda is a small anonymous function defined in one line: `lambda args: expression`

```python
square = lambda x: x ** 2   # define a one-line function and assign it to a variable
print(square(5))             # 25 — call it just like a regular function

add = lambda a, b: a + b    # lambdas can take multiple parameters
print(add(3, 4))             # 7
```

## With map(), filter(), sorted()

```python
nums = [1, 2, 3, 4, 5]

# map — apply a function to every element and return a new iterable
squares = list(map(lambda x: x**2, nums))  # wrap in list() to see the results
print(squares)  # [1, 4, 9, 16, 25]

# filter — keep only elements where the function returns True
evens = list(filter(lambda x: x % 2 == 0, nums))  # only even numbers pass the test
print(evens)  # [2, 4]

# sorted with key — use a lambda to define what to sort by
words = ['banana', 'apple', 'cherry']
print(sorted(words, key=lambda w: len(w)))  # sort by word length, shortest first
```
