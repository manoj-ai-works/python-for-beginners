# *args and **kwargs
`*args` collects extra positional arguments. `**kwargs` collects extra keyword arguments.

## 1. *args

```python
def total(*args):       # *args collects any number of positional arguments into a tuple
    return sum(args)    # 'args' is a tuple, so we can pass it to sum()

print(total(1, 2, 3))        # 6   — three arguments
print(total(10, 20, 30, 40)) # 100 — four arguments; *args handles any count
```

## 2. **kwargs

```python
def profile(**kwargs):              # **kwargs collects keyword arguments into a dictionary
    for key, value in kwargs.items():  # iterate over the key-value pairs
        print(f"{key}: {value}")

profile(name="Alice", age=25, city="NYC")  # each keyword becomes a key in the dict
```

## 3. Combining both

```python
def demo(a, b, *args, **kwargs):  # regular params first, then *args, then **kwargs
    print(a, b)       # a=1, b=2 — captured by regular parameters
    print(args)       # (3, 4)   — extra positional args go into the tuple
    print(kwargs)     # {'x': 10, 'y': 20} — keyword args go into the dict

demo(1, 2, 3, 4, x=10, y=20)
```
