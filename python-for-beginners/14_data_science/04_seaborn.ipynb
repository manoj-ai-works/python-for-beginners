# Seaborn
Seaborn is built on top of Matplotlib and makes statistical visualizations much easier. It works great with Pandas DataFrames.

Install: `pip install seaborn`

## 1. Setup and Sample Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style='whitegrid')  # apply a clean default theme to all plots

# Seaborn ships with built-in datasets — great for learning
tips = sns.load_dataset('tips')   # restaurant tips data
print(tips.head())
print(tips.shape)                 # (244, 7)
```

## 2. Distribution Plots

```python
# histplot — histogram with optional KDE (kernel density estimate) curve
plt.figure(figsize=(7, 4))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='steelblue')
# kde=True overlays a smooth density curve on top of the histogram
plt.title('Distribution of Total Bill')
plt.tight_layout()
plt.show()

# kdeplot — just the smooth density curve, no bars
plt.figure(figsize=(7, 4))
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True, alpha=0.4)
# hue='sex' draws a separate curve for each gender
plt.title('Bill Distribution by Gender')
plt.tight_layout()
plt.show()
```

## 3. Categorical Plots

```python
# boxplot — shows median, quartiles, and outliers
plt.figure(figsize=(7, 4))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='Set2')
plt.title('Total Bill by Day and Gender')
plt.tight_layout()
plt.show()

# violinplot — like a boxplot but shows the full distribution shape
plt.figure(figsize=(7, 4))
sns.violinplot(data=tips, x='day', y='tip', palette='muted')
plt.title('Tip Distribution by Day')
plt.tight_layout()
plt.show()

# barplot — shows mean with confidence interval error bars
plt.figure(figsize=(7, 4))
sns.barplot(data=tips, x='day', y='tip', hue='sex', palette='Set1')
plt.title('Average Tip by Day and Gender')
plt.tight_layout()
plt.show()
```

## 4. Scatter and Regression Plots

```python
# scatterplot — with hue and size encoding
plt.figure(figsize=(7, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', size='size',
                sizes=(40, 200), alpha=0.7)
# hue encodes gender as color, size encodes party size as dot size
plt.title('Tip vs Total Bill')
plt.tight_layout()
plt.show()

# regplot — scatter + linear regression line
plt.figure(figsize=(7, 5))
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha': 0.5})
# regplot automatically fits and draws a regression line with confidence band
plt.title('Tip vs Total Bill with Regression Line')
plt.tight_layout()
plt.show()
```

## 5. Heatmap — Correlation Matrix

```python
# Correlation matrix shows how strongly each pair of numeric columns is related
# Values range from -1 (perfect negative) to +1 (perfect positive)
corr = tips[['total_bill', 'tip', 'size']].corr()

plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
# annot=True writes the correlation value inside each cell
# cmap='coolwarm' — blue for negative, red for positive correlation
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 6. pairplot — All Relationships at Once

```python
# pairplot creates a grid of scatter plots for every pair of numeric columns
# Diagonal shows the distribution of each individual column
sns.pairplot(tips[['total_bill', 'tip', 'size']], diag_kind='kde')
plt.suptitle('Pairplot of Tips Dataset', y=1.02)
plt.show()
```
