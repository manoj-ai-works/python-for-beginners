# Matplotlib
Matplotlib is Python's core plotting library. It gives you full control over every element of a chart.

Install: `pip install matplotlib`

## 1. Basic Line Plot

```python
import matplotlib.pyplot as plt  # 'plt' is the universal alias
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)

plt.figure(figsize=(8, 4))           # set figure size in inches (width, height)
plt.plot(x, y, color='steelblue', linewidth=2, label='sin(x)')  # draw the line
plt.plot(x, np.cos(x), color='tomato', linestyle='--', label='cos(x)')  # second line

plt.title('Sine and Cosine')         # chart title
plt.xlabel('x')                      # x-axis label
plt.ylabel('y')                      # y-axis label
plt.legend()                         # show the legend (uses 'label' from plot())
plt.grid(True, alpha=0.3)            # light grid lines
plt.tight_layout()                   # prevent labels from being cut off
plt.show()                           # display the chart
```

## 2. Bar Chart

```python
languages = ['Python', 'JavaScript', 'Java', 'C++', 'Rust']
popularity = [30, 25, 18, 12, 8]  # arbitrary popularity scores

plt.figure(figsize=(8, 5))
bars = plt.bar(languages, popularity, color='steelblue', edgecolor='white')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.3,
             str(height), ha='center', va='bottom', fontsize=10)

plt.title('Programming Language Popularity')
plt.ylabel('Popularity Score')
plt.tight_layout()
plt.show()
```

## 3. Scatter Plot

```python
np.random.seed(42)  # seed for reproducibility
study_hours = np.random.uniform(1, 10, 50)                    # 50 random study hours
scores      = study_hours * 8 + np.random.normal(0, 5, 50)   # score roughly = hours * 8 + noise

plt.figure(figsize=(7, 5))
plt.scatter(study_hours, scores, color='steelblue', alpha=0.7, edgecolors='white')
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Score')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. Histogram

```python
data = np.random.normal(loc=70, scale=10, size=500)  # 500 scores, mean=70, std=10

plt.figure(figsize=(7, 4))
plt.hist(data, bins=20, color='steelblue', edgecolor='white', alpha=0.8)
plt.axvline(data.mean(), color='tomato', linestyle='--', label=f'Mean: {data.mean():.1f}')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

## 5. Subplots — Multiple Charts in One Figure

```python
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns of subplots

# Left subplot
axes[0].plot(x, np.sin(x), color='steelblue')
axes[0].set_title('Sine Wave')
axes[0].set_xlabel('x')
axes[0].grid(True, alpha=0.3)

# Right subplot
axes[1].plot(x, x ** 2, color='tomato')
axes[1].set_title('Quadratic')
axes[1].set_xlabel('x')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
