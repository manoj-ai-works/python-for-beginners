# Pandas
Pandas is the go-to library for working with tabular data — think spreadsheets or SQL tables in Python.

Install: `pip install pandas`

## 1. Series and DataFrame

```python
import pandas as pd  # 'pd' is the universal alias

# Series — a 1D labeled array (like a single column)
scores = pd.Series([85, 92, 78, 95, 88], name='score')
print(scores)
print(scores.mean())   # 87.6

# DataFrame — a 2D table with labeled rows and columns
data = {
    'name':  ['Alice', 'Bob', 'Carol', 'Dave'],
    'age':   [25, 30, 22, 35],
    'score': [85, 92, 78, 95],
    'city':  ['NYC', 'LA', 'NYC', 'Chicago'],
}
df = pd.DataFrame(data)
print(df)
print(df.shape)        # (4, 4) — 4 rows, 4 columns
print(df.dtypes)       # data type of each column
```

## 2. Exploring Data

```python
print(df.head(2))          # first 2 rows
print(df.tail(2))          # last 2 rows
print(df.info())           # column names, types, non-null counts
print(df.describe())       # count, mean, std, min, max for numeric columns

# Accessing columns
print(df['name'])          # single column → Series
print(df[['name', 'score']]) # multiple columns → DataFrame

# Accessing rows
print(df.iloc[0])          # row by integer position (0-indexed)
print(df.loc[0])           # row by label (same here since index is 0,1,2,3)
print(df.iloc[1:3])        # rows 1 and 2
```

## 3. Filtering and Sorting

```python
# Boolean filtering — like numpy boolean indexing
high_scorers = df[df['score'] > 85]          # rows where score > 85
print(high_scorers)

nyc_high = df[(df['city'] == 'NYC') & (df['score'] > 80)]  # combine conditions with &
print(nyc_high)

# Sorting
print(df.sort_values('score', ascending=False))  # sort by score, highest first
print(df.sort_values(['city', 'score']))          # sort by city, then score
```

## 4. Adding, Modifying, and Dropping Columns

```python
# Add a new column — computed from existing ones
df['grade'] = df['score'].apply(lambda s: 'A' if s >= 90 else 'B' if s >= 80 else 'C')
print(df)

# Modify a column
df['age'] = df['age'] + 1   # everyone gets a year older

# Drop a column
df_clean = df.drop(columns=['grade'])  # returns a new DataFrame, doesn't modify df
print(df_clean.columns)
```

## 5. Grouping and Aggregation

```python
# groupby — split data into groups, apply a function, combine results
city_avg = df.groupby('city')['score'].mean()  # average score per city
print(city_avg)

# Multiple aggregations at once
summary = df.groupby('city')['score'].agg(['mean', 'max', 'count'])
print(summary)
```

## 6. Handling Missing Data

```python
import numpy as np

df2 = pd.DataFrame({
    'name':  ['Alice', 'Bob', 'Carol'],
    'score': [85, np.nan, 78],   # np.nan represents a missing value
    'city':  ['NYC', None, 'LA'],
})

print(df2.isnull())              # True where value is missing
print(df2.isnull().sum())        # count of missing values per column

df2_filled = df2.fillna({'score': df2['score'].mean(), 'city': 'Unknown'})  # fill with defaults
df2_dropped = df2.dropna()       # drop any row that has at least one missing value

print(df2_filled)
```

## 7. Reading and Writing Files

```python
# Save to CSV
df.to_csv('students.csv', index=False)  # index=False avoids writing the row numbers

# Read from CSV
df_loaded = pd.read_csv('students.csv')
print(df_loaded.head())

# Also works with Excel, JSON, SQL...
# df.to_excel('students.xlsx', index=False)
# df.to_json('students.json', orient='records')
```
